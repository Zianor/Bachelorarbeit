```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

from data_preparation import Data
import utils
```

```python
data = pd.read_csv(utils.get_features_csv_path(segment_length=10, overlap_amount=0, hr_threshold=10), index_col=False)
```

```python
%matplotlib notebook
```

```python
for patient_id in data['patient_id'].unique():
    fig, ax = plt.subplots(figsize=(10,10))

    data_patient = data[data['patient_id']==patient_id]
    
    plt.plot(data_patient.index[:], data_patient['ecg_hr'][:], label='ecg')
    plt.plot(data_patient.index[:], data_patient['bcg_hr'][:], label='bcg')
    plt.title("Patient " + str(patient_id))
    ax.legend()

    plt.show()
```
