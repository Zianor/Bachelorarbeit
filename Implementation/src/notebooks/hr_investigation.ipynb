```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

from src.ml_statistical_features import load_data_as_dataframe
from src.data_preparation import Data
```

```python
data = load_data_as_dataframe()
```

```python
%matplotlib notebook
```

```python
for patient_id in data['patient_id'].unique():
    fig, ax = plt.subplots(figsize=(10,10))

    data_patient = data[data['patient_id']==patient_id]
    
    plt.plot(data_patient.index, data_patient['ecg_hr'], label='mean')
    plt.plot(data_patient.index, data_patient['ecg_hr_std'], label='std')
    plt.plot(data_patient.index, data_patient['bcg_hr'], label='bcg')
    ax.legend()

    plt.show()
```
